```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv", header=1)
```

```python
dataset
```

```python
dataset.info()
```

data cleaning

```python
#missing values
dataset[dataset.isnull().any(axis=1)]
```

```python
df=dataset
df.loc[:122,"region"]=0
df.loc[122:,"region"]=1
```

```python
df.info()
```

```python
df[['region']]=df[['region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
##removing the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
#dropping the 122 row
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fixing spaces in column names
df.columns= df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
#change the required columns as an integer data type
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
# changing the columns to float datatype
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
# lets save the cleaned data set
df.to_csv('Algerian_forest_fires_cleaned.csv',index=False)
```
